```python
#Tamim Mahmud from Daffodil International University
#22/06/2025
import torch
import matplotlib.pyplot as plt

# Step 1: Initial weight
w = 0.0

# Step 2: Learning rate
lr = 0.1

# Step 3: Number of steps
epochs = 25

# Step 4: Store history
loss_list = []
w_list = []

# Step 5: Loop for manual gradient descent
for i in range(epochs):
    # Function: f(w) = (2w - 4)^2
    # Derivative: f'(w) = 2 * (2w - 4) * 2 = 8w - 16
    grad = 8 * w - 16

    # Update weight
    w = w - lr * grad

    # Compute current loss
    loss = (2 * w - 4) ** 2

    # Track
    loss_list.append(loss)
    w_list.append(w)

    print(f"Epoch {i+1}: w = {w:.4f}, loss = {loss:.4f}")

# Plotting loss
plt.plot(loss_list)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Manual Gradient Descent Loss")
plt.grid(True)
plt.show()
```

```python
#Gradient descent with PyTorch

import torch
w=torch.tensor([0.0],requires_grad=True)
optimizer=torch.optim.SGD([w],lr=0.1)

for i in range(10):
  optimizer.zero_grad()

  loss=(2*w-4)**2
  loss.backward()
  optimizer.step()
  print(f"Epoch{i+1} : W= {w.item():.4f} ,Loss= {loss.item():.4f}")
```

```python
#with linear regression
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Step 1: Training data (x, y)
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[3.0], [5.0], [7.0], [9.0]])  # y = 2x + 1

# Step 2: Model - simple linear model: y = wx + b
model = nn.Linear(in_features=1, out_features=1)

# Step 3: Loss function
loss_fn = nn.MSELoss()

# Step 4: Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Step 5: Training loop
epochs = 100
loss_history = []

for epoch in range(epochs):
    # Forward pass
    y_pred = model(x)

    # Compute loss
    loss = loss_fn(y_pred, y)

    # Backpropagation
    optimizer.zero_grad()
    loss.backward()

    # Gradient descent step
    optimizer.step()

    # Save loss
    loss_history.append(loss.item())

    # Print
    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")

# Step 6: Show final parameters
w, b = model.parameters()
print("\nLearned weight:", w.item())
print("Learned bias:", b.item())

# Step 7: Plot loss curve
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.grid(True)
plt.show()
```
